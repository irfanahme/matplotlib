```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# function method 
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object origent method 

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
#axes.plot(x,y)
#plt.tight_layout()
#for current_ax in axes:
   # current_ax.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

## Figure Size and DPI

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X squared')
ax.plot(x,x**3, label = 'X cubed')
ax.legend(loc = (0.1,0.1))
```

```python
fig.savefig('my_picture.jpg', dpi = 200)
```

# Plot Appearance 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'purple', lw = 3,  ls = '-', marker = '*', markersize = 20,
       markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')
```
